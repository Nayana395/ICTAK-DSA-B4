# Scatter Plot

```python
plt.figure(figsize=(5,3.5))
# c - color
plt.scatter(data["weight"], data["mpg"], c="k", marker="*")
plt.title("MPG vs Weight")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

# Sub-Figure

```python
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(data["weight"],data["mpg"], c="k")
plt.ylabel("MPG")
plt.xlabel("Weight")
plt.show()
plt.subplot(2,1,2)
plt.scatter(data["displacement"],data["mpg"], c="red")
plt.ylabel("MPG")
plt.xlabel("Displacement")
plt.show()
```

```python
# Feature Binning
from scipy import stats

data = np.random.exponential(scale=1,size=1000)
transformed_data, lambda_value = stats.boxcox(data)

plt.figure(figsize=(8,5))

plt.subplot(1,2,1)
plt.hist(data, bins=30)
plt.title("Original Data")

plt.subplot(1,2,2)
plt.hist(transformed_data, bins=30)
plt.title("Box-Cox Data")

plt.show()
```
