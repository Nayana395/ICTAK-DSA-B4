# Diabetes Dataset EDA:

````python
df_diabetes = pd.read_csv("diabetes_model.csv")
df_diabetes
````

## Univariate Analysis:
- Analyze individual variables (e.g., histograms, box plots).

````python
# histogram of count of pregnancies
sns.histplot(df_diabetes["Pregnancies"])
````

````python
# Box plot of all the BMIs
sns.boxplot(df_diabetes["BMI"])
````

## Bivariate Analysis:
- Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices).

````python
# Relationship between blood pressure and age
plt.scatter(df_diabetes["Age"],df_diabetes["BloodPressure"])
````

````python
print("Pearson Correlation Coefficient = \n", np.corrcoef(df_diabetes["Age"], df_diabetes["BloodPressure"]))
````
