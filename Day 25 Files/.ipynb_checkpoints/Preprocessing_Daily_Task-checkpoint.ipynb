```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Datasets already available inside seaborn
# sns.get_dataset_names()

data = sns.load_dataset('titanic')
data
```

```python
data.describe()
```

```python
# Converting yes/no in 'alive' column to binary
data['alive'].replace({'yes':1,'no':0},inplace=True)
data['alive'] = data['alive'].astype('int64')
```

```python
data['class'].unique()
```

```python
# Converting textual info in 'class' column to numerical values
data['class'].replace({'First':1,'Second':2, 'Third':3},inplace=True)
data['class'] = data['class'].astype('int64')
```

```python
data.dtypes
```

```python
# Dropping duplicate columns
data = data.iloc[:,2:]
```

```python
# Dropping duplicate rows
data.drop_duplicates(inplace=True)
# dup_rows = data.duplicated()
# data = data[~dup_rows]
```

```python
data
```

```python
# count of missing values
data.isna().sum()
```

```python
data.dtypes
```

```python
num_data = data.select_dtypes(include="number")
cat_data = data.select_dtypes(exclude="number")
```

```python
print("Numerical columns are: ")
num_cols = num_data.columns.tolist()
print(num_cols)

print("\nCategorical columns are: ")
cat_cols = cat_data.columns.tolist()
print(cat_cols)
```

```python
for col in num_cols:
    plt.hist(data[col])
    plt.title("Histogram of {} column".format(col))
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

```python
for col in num_cols:
    if (col=='alive' or col=='class'):
        data[col] = data[col].fillna(data[col].mode())
    elif (col=='age'):
        data[col] = data[col].fillna(data[col].mean())
    else:
        data[col] = data[col].fillna(data[col].median())
```

```python
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
data.isna().sum()
```

```python
plt.boxplot(num_data)
plt.xticks([1, 2, 3, 4,5,6], num_cols, rotation=45)
plt.show()
```

```python
def remove_outliers(df, column_name):
    q1 = df[column_name].quantile(0.25)
    q3 = df[column_name].quantile(0.75)
    iqr = q3-q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    df[column_name] = df[column_name].clip(upper=upper_bound)
    df[column_name] = df[column_name].clip(lower=lower_bound)
    return df[column_name]
```

```python
for col in num_cols:
    data[col] = remove_outliers(data, col)
```

```python
for col in num_cols:
    num_data[col] = remove_outliers(num_data, col)
```

```python
plt.boxplot(num_data)
plt.xticks([1, 2, 3, 4,5,6], num_cols, rotation=45)
plt.show()
```
